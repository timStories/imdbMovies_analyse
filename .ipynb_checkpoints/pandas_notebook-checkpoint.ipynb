# TMDb Films with over 50 votes

The purpose of this analysis is to find important films of every genre.

Data set:
https://www.kaggle.com/datasets/asaniczka/tmdb-movies-dataset-2023-930k-movies

This was the base data set that I have made smaller to only include films where vote_count > 50. This is done to ensure to include only films that are deemed 'relevant' by IMDb users. I have created the smaller csv by using HDFS and Spark.

# Genres

```python
unique_genres = df['genres'].unique()
count_genres = df['genres'].value_counts()
print(unique_genres)
print(count_genres)
```

```python
# Step 1: Split the 'genres' column
df['genres'] = df['genres'].str.split(', ')

# Step 2: Explode the DataFrame
df_exploded = df.explode('genres')

# Step 3: Create separate DataFrames for each genre
genre_dfs = {genre: df_exploded[df_exploded['genres'] == genre] for genre in df_exploded['genres'].unique()}
```

```python
# Create a frequency table of genres
genre_counts = df_exploded['genres'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

print(genre_counts)
```

```python
# Specify the genre you want to display
genre_to_display = 'War'

# Filter the DataFrame for the specified genre
genre_df = df_exploded[df_exploded['genres'] == genre_to_display]

# Display the DataFrame for the specified genre
genre_df.head()
```

## Single genre selector

```python
# 1. Create the genre_film_dict
genre_film_dict = df_exploded.groupby('genres')['title'].apply(list).to_dict()

# 2. (Optional) Create a genre_counts dictionary (if you need it separately)
genre_counts_dict = df_exploded['genres'].value_counts().to_dict()

genre_counts_dict
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='genre-dropdown',
        options=[
            {'label': f"{genre} ({count})", 'value': genre}
            for genre, count in genre_counts_dict.items()  # Use genre_counts_dict here
        ],
        multi=True  # Allow multiple genre selections
    ),
    html.Div(id='selected-films')
])

@app.callback(
    Output('selected-films', 'children'),
    Input('genre-dropdown', 'value')
)
def update_films(selected_genres):
    if selected_genres is None or not selected_genres:
        return "Select genres to see films."

    selected_films = set()

    for genre in selected_genres:
        films = genre_film_dict.get(genre, [])  # Use genre_film_dict here
        selected_films.update(films)

    if not selected_films:
        return "No films match the selected genres."

    return html.Ul([html.Li(film) for film in selected_films])

if __name__ == '__main__':
    app.run_server(debug=True, localhost=8050)
```

## Single keyword selector

```python
# 1. Create the keyword_film_dict (same as before)
keyword_film_dict = df_exploded.groupby('keywords_list')['title'].apply(list).to_dict()

# 2. Filter for frequent keywords (appearing 5 or more times)
frequent_keywords = keyword_counts[keyword_counts >= 5]

# 3. Filter the keyword_film_dict to include only frequent keywords AND get counts
frequent_keyword_film_dict = {}
for keyword, films in keyword_film_dict.items():
    if keyword in frequent_keywords:
        frequent_keyword_film_dict[keyword] = films  # Keep films
        
# Get counts for the frequent keywords
frequent_keyword_counts = {keyword: len(films) for keyword, films in frequent_keyword_film_dict.items()}
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import pandas as pd
import numpy as np

app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Dropdown(
        id='keyword-dropdown',
        options=[
            {'label': f"{k} ({v})", 'value': k}  # Include counts in the dropdown labels
            for k, v in frequent_keyword_counts.items()
        ],
        multi=True
    ),
    html.Div(id='selected-films')
])

@app.callback(
    Output('selected-films', 'children'),
    Input('keyword-dropdown', 'value')
)
def update_films(selected_keywords):
    if selected_keywords is None or not selected_keywords:
        return "Select keywords to see films."

    selected_films = set()

    for keyword in selected_keywords:
        films = frequent_keyword_film_dict.get(keyword, [])  # Use the correct dictionary
        selected_films.update(films)

    if not selected_films:
        return "No films match the selected keywords."

    return html.Ul([html.Li(film) for film in selected_films])

if __name__ == '__main__':
    app.run_server(debug=True, localhost=8051)
```

# Keywords + Genre Selector

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output

app = dash.Dash(__name__)

app.layout = html.Div([
    html.H3("Filter Films"),  # Title for the section

    html.Div([  # Container for the genre dropdown (GENRES FIRST)
        html.Label("Genres:"),
        dcc.Dropdown(
            id='genre-dropdown',
            options=[
                {'label': f"{genre} ({count})", 'value': genre}
                for genre, count in genre_counts_dict.items()
            ],
            multi=True
        )
    ]),

    html.Div([  # Container for the keyword dropdown (KEYWORDS SECOND)
        html.Label("Keywords:"),
        dcc.Dropdown(
            id='keyword-dropdown',
            options=[
                {'label': f"{k} ({v})", 'value': k}
                for k, v in frequent_keyword_counts.items()
            ],
            multi=True
        )
    ]),

    html.Div(id='selected-films')
])

@app.callback(
    Output('selected-films', 'children'),
    Input('genre-dropdown', 'value'),  # Input order swapped
    Input('keyword-dropdown', 'value')
)
def update_films(selected_genres, selected_keywords):  # Parameter order swapped
    # ... (rest of the callback logic remains exactly the same)
    if (selected_keywords is None or not selected_keywords) and \
            (selected_genres is None or not selected_genres):
        return "Select genres and/or keywords to see films."

    selected_films = set()

    # Filter by genres (GENRES FIRST)
    if selected_genres:
        films_by_genre = set()
        for genre in selected_genres:
            films = genre_film_dict.get(genre, [])
            films_by_genre.update(films)
    else:
        films_by_genre = None

    # Filter by keywords (KEYWORDS SECOND)
    if selected_keywords:
        films_by_keyword = set()
        for keyword in selected_keywords:
            films = frequent_keyword_film_dict.get(keyword, [])
            films_by_keyword.update(films)
    else:
        films_by_keyword = None

    # Combine filters
    if films_by_keyword is not None and films_by_genre is not None:
        selected_films = films_by_keyword.intersection(films_by_genre)
    elif films_by_genre is not None:  # Genre filter first
        selected_films = films_by_genre
    elif films_by_keyword is not None:  # Then keyword filter
        selected_films = films_by_keyword
    else:
        selected_films = set()

    if not selected_films:
        return "No films match the selected criteria."

    return html.Ul([html.Li(film) for film in selected_films])

if __name__ == '__main__':
    app.run_server(debug=True, localhost=8055)
```
