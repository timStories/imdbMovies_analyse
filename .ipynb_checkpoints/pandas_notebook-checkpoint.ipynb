# Genres

```python
unique_genres = df['genres'].unique()
count_genres = df['genres'].value_counts()
print(unique_genres)
print(count_genres)
```

```python
# Step 1: Split the 'genres' column
df['genres'] = df['genres'].str.split(', ')

# Step 2: Explode the DataFrame
df_exploded = df.explode('genres')

# Step 3: Create separate DataFrames for each genre
genre_dfs = {genre: df_exploded[df_exploded['genres'] == genre] for genre in df_exploded['genres'].unique()}
```

```python
# Create a frequency table of genres
genre_counts = df_exploded['genres'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

print(genre_counts)
```

```python
# Specify the genre you want to display
genre_to_display = 'War'

# Filter the DataFrame for the specified genre
genre_df = df_exploded[df_exploded['genres'] == genre_to_display]

# Display the DataFrame for the specified genre
genre_df.head()
```

# Keywords

```python
# Assuming your DataFrame is named 'df'
df_test = df.head(10)
```

```python
import pandas as pd
import numpy as np

# 1. Convert the 'keywords' column to string type (if it's not already)
df_test['keywords'] = df_test['keywords'].astype(str)

# 2. Split the keywords by ', ' and create a list of lists
df_test['keywords_list'] = df_test['keywords'].apply(lambda x: x.split(', '))

# 3. Explode the list of lists to have one keyword per row
df_exploded = df_test.explode('keywords_list')

# 4. Get all unique keywords (using a set for uniqueness and removing NaNs)
all_keywords = df_exploded['keywords_list'].dropna().unique()

# 5. Corrected recursive flattening function
def flatten_and_extract(arr):
    result = []
    for item in arr:
        if isinstance(item, (np.ndarray, list)):  # Check for both ndarray and list
            result.extend(flatten_and_extract(item))
        elif isinstance(item, str):
            result.append(item)
    return result

# Flatten and extract the keywords
cleaned_keywords = flatten_and_extract(all_keywords)

# 6. Convert to a standard Python list and remove duplicates
cleaned_keywords = list(set(cleaned_keywords))  # Convert to set to remove duplicates

# 7. Print the cleaned keywords
for keyword in cleaned_keywords:
    print(keyword)
```
